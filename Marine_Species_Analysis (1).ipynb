```python
numeric_cols = ['LAT', 'LON', 'Depth', 'Year', 'wtcpue']
for df in [muellersSeaPansy, brownRockShrimp, paperScallop, emeraldParrotfish, shark]:
    df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python

import numpy as np

predator_df = shark.copy()
mueller_df = muellersSeaPansy.copy()
shrimp_df = brownRockShrimp.copy()
scallop_df = paperScallop.copy()
parrotfish_df = emeraldParrotfish.copy()

def find_nearby_prey(predator_df, prey_df, lat_diff=0.5, lon_diff=0.5, year_diff=1):
    matched_prey = []


    for _, predator_row in predator_df.iterrows():

        nearby_prey = prey_df[
            (np.abs(prey_df['LAT'] - predator_row['LAT']) <= lat_diff) &
            (np.abs(prey_df['LON'] - predator_row['LON']) <= lon_diff) &
            (np.abs(prey_df['Year'] - predator_row['Year']) <= year_diff)
        ]

        if not nearby_prey.empty:

            avg_prey_lat = nearby_prey['LAT'].mean()
        else:
            avg_prey_lat = np.nan

        matched_prey.append(avg_prey_lat)

    return matched_prey


predator_df['Nearby_Mueller_LAT'] = find_nearby_prey(predator_df, mueller_df)
predator_df['Nearby_Shrimp_LAT'] = find_nearby_prey(predator_df, shrimp_df)
predator_df['Nearby_Scallop_LAT'] = find_nearby_prey(predator_df, scallop_df)
predator_df['Nearby_Parrotfish_LAT'] = find_nearby_prey(predator_df, parrotfish_df)


predator_df.head()
```

```python
from sklearn.linear_model import LinearRegression

X = predator_df[['Nearby_Mueller_LAT', 'Nearby_Shrimp_LAT', 'Nearby_Scallop_LAT', 'Nearby_Parrotfish_LAT']].fillna(0)
y = predator_df['LAT'].astype(float)

model = LinearRegression().fit(X, y)


print('Regression coefficients:', model.coef_)
print('Intercept:', model.intercept_)
print('R-squared score:', model.score(X, y))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 4, figsize=(18, 5))


sns.regplot(x='Nearby_Mueller_LAT', y='LAT', data=predator_df, ax=axes[0])
axes[0].set_title("Predator Latitude vs Mueller's Sea Pansy Latitude")
axes[0].set_xlabel("Mueller's Sea Pansy Latitude")
axes[0].set_ylabel("Predator Latitude")


sns.regplot(x='Nearby_Shrimp_LAT', y='LAT', data=predator_df, ax=axes[1])
axes[1].set_title("Predator Latitude vs Brown Rock Shrimp Latitude")
axes[1].set_xlabel("Brown Rock Shrimp Latitude")
axes[1].set_ylabel("Predator Latitude")


sns.regplot(x='Nearby_Scallop_LAT', y='LAT', data=predator_df, ax=axes[2])
axes[2].set_title("Predator Latitude vs Paper Scallop Latitude")
axes[2].set_xlabel("Paper Scallop Latitude")
axes[2].set_ylabel("Predator Latitude")


sns.regplot(x='Nearby_Parrotfish_LAT', y='LAT', data=predator_df, ax=axes[3])
axes[3].set_title("Predator Latitude vs Emerald Parrotfish Latitude")
axes[3].set_xlabel("Emerald Parrotfish Latitude")
axes[3].set_ylabel("Predator Latitude")


plt.tight_layout()
plt.show()
```

```python

sns.pairplot(predator_df[['LAT', 'Nearby_Mueller_LAT', 'Nearby_Shrimp_LAT', 'Nearby_Scallop_LAT', 'Nearby_Parrotfish_LAT']])
plt.show()
```

```python

corr = predator_df[['LAT', 'Nearby_Mueller_LAT', 'Nearby_Shrimp_LAT', 'Nearby_Scallop_LAT', 'Nearby_Parrotfish_LAT']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix between Predator and Prey Latitudes')
plt.show()
```
